## We will go through process step 2. Exploratory Dta Analysis(EDA) - Univariate <br><br>
### It means Independent analysis of each columns in dataset <br><br>
### a. identifying and differentiating the dataset in Numerical and Categorical Data &nbsp; <br> <br> b. Categoriacal Data Representation- countplot, piechart, etc., <br> <br>c. Numerical Data Representation- histogram, distplot, boxplot, etc.,

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('medical_students_dataset.csv')
```

```python
df.sample(10)
```

```python
df.info()
```

## a. identifying and differentiating the dataset in Numerical and Categorical Data  

### Numerical data features(colums) - <br> Age <br> Height <br> Weight <br> BMI <br> Temperature <br> Heart Rate <br> Blood Pressure <br> Cholesterol <br> <br>
### Categorical data features(columns) - <br> Gender <br> Blood Type <br> Diabetes <br> Smoking

## b. Categoriacal Data Representation- 
## a.countplot <br>b.piechart<br>c.etc.,

## 1.Gender 

```python
df['Gender'] = df['Gender'].map({'Male':0 ,'Female':1})
```

```python
df['Gender'] = df['Gender'].astype('category')
sns.countplot(df,x='Gender')
#you can use below also fro ploting 
# df['Gender'].value_counts().plot(kind='bar')
df['Gender'].value_counts()
```

```python
df['Gender'].value_counts().plot(kind='pie',autopct='%1.0f%%')
```

## 2.Blood Type

```python
df['Blood Type'].value_counts()
```

```python
df['Blood Type'] = df['Blood Type'].map({'A':1 ,'B':2,'O':3,'AB':4})
```

```python
df['Blood Type'].value_counts()
```

```python
sns.countplot(df,x='Blood Type')
```

```python
df['Blood Type'].value_counts().plot(kind='pie', autopct='%1.0f%%')
```

## 3.Diabetes

```python
df['Diabetes'] = df['Diabetes'].map({'Yes':1, 'No':0})
sns.countplot(df, x= 'Diabetes')
df['Diabetes'].value_counts()
```

```python
df['Diabetes'].value_counts().plot(kind='pie', autopct= '%1.0f%%')
```

## 4. Smoking

```python
df['Smoking'] = df['Smoking'].map({'Yes':1,'No':0})
sns.countplot(df, x = 'Smoking')
df['Smoking'].value_counts()
```

```python
df['Smoking'].value_counts().plot(kind='pie', autopct='%1.0f%%')
```
