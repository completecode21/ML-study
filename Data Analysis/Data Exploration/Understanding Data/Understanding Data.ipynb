### d. Knowing the data type of the columns

```python
#we can see how much memory is used by the data and datatypes of the features
df.info()
```

```python
#here Age column data type is null which is unnesessory so replace it as int so we can reduce some sapce
df['Age'] = df['Age'].astype(int)
df['Years of Experience'] = df['Years of Experience'].astype(int)
df['Salary'] = df['Salary'].astype(int)
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.info()
```

####  with chnaging datatypes and removing extra columns we have saved space upto 132kb it is very impressive in compare to size of dataset
